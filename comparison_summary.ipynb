```python
#Visualize Linear Equation

# Linear Regression Equation:
# y = β0 + β1x1 + β2x2 + ... + βnxn
# Output: Continuous value
```

```python
#Visualize Sigmoid Function

import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

z = np.linspace(-10, 10, 100)
y = sigmoid(z)

plt.figure(figsize=(6,5))
plt.plot(z, y)
plt.title("Sigmoid Function (Used in Logistic Regression)")
plt.xlabel("z")
plt.ylabel("Probability")
plt.show()
```

```python
#Logistic Regression transforms linear output into probability using the Sigmoid function.
```

```python
#Clear Comparison Table

#|     Linear Regression       |      Logistic Regression       |
#|-----------------------------|--------------------------------|
#| Used for Regression         | Used for Classification        |
#| Outputs Continuous Values   | Outputs Probability (0–1)      |
#| Uses Linear Equation        | Uses Linear Equation + Sigmoid |
#| Can predict any real number | Predicts Class Label           |
```

```python
#Linear vs Logistic Comparison
x = np.linspace(-10, 10, 200)

# Linear function
linear = x

# Sigmoid function
sigmoid = 1 / (1 + np.exp(-x))

plt.figure(figsize=(7,5))
plt.plot(x, linear)
plt.plot(x, sigmoid)

plt.xlabel("Input Value")
plt.ylabel("Output")
plt.title("Linear vs Logistic Regression Output")
plt.legend(["Linear Output", "Logistic (Sigmoid) Output"])
plt.show()
```
